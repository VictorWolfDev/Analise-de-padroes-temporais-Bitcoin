```python
### Agosto é realmente um mês "fraco"?

Sim, Gráficos apontam que agosto tende a ser mais fraco em comparação a outros meses por ser um mês de baixa liquidez e volume de 
negociação, tornando o bitcoin e outras criptomoedas como ethereum mais voláteis e propensas a cair com noticias.
Um dos fatores para que isso ocorra são as férias de verão no hemisfério norte, onde os investidores e traders vendem seus ativos 
em maio e junho para aproveitarem os lúcros e não se preocuparem com a volatilidade do mercado. Esse evento é também chamado de
"Sell in may and go away".
Outros fatores importantes são fatores macroeconômicos e eventos do próprio mercado de criptomoedas.

## Meses com mais variações negativas entre 2010 e 2024 (Acima de 6 ocorrências):
            
Março - 7 ocorrências
Agosto - 9 ocorrências
Setembro - 8 ocorrências
Dezembro - 7 ocorrências

## Meses com quedas coincidentes a agosto anual
     ________________________________________
    |  MESES  |  OCORRÊNCIAS COINCIDENTES    |
    |---------|------------------------------|
    |Março    | 2011, 2014, 2018             |
    |Agosto   | 2011, 2014, 2018, 2019, 2022 |
    |Setembro | 2011, 2014, 2018, 2019, 2022 |
    |Dezembro |       2014, 2018, 2019, 2022 |
     ----------------------------------------

    *A tabela apresenta anos em que os meses com mais ocorrências de quedas teve quedas em anos coincidentes a agosto com
    no minimo duas vezes.*

## Uso de indicadores on-chain para análise
Dados on-chain como MVRV-Z, RHODL Ratio, e SOPR(Spent Output Profit Ratio) não indicam situações anormais no mês de agosto 
como alta de investidores de curto prazo e consequentemente queda por FUD após FOMO, grandes prejuizos ou supervalorização,
o que pode levar a entender que é um movimento que ocorre por fatores culturais ou eventos sazonais regionais em paises com 
maiores negociações de Bitcoin ou eventos pontuais macroeconômicos e ciclico do mercado de criptomoedas.

    
## Questões para investigação

**Quais foram os anos de queda em agosto?**
    **R:** Agosto - 2024, 2023, 2022, 2019, 2018, 2016, 2015, 2014, 2011


**Quão grande é a queda em agosto e setembro comparado a queda em outros meses? devemos nos preocuparmos?**

    **R:** A queda em agosto em média não costuma ser grande comparado a outros meses com mal desempenho históricamente,
tendo uma variação de apenas −0,66% a −2,26%.
A menos que seja ano de bearmarket, não há motivos para se preocupar com a queda em agosto visto que costuma 
ser uma correção saúdavel. O histórico mostra que em bearmakets, o preço pode alcançar patamares de até -38% de 
queda.

  
    
### Gatilhos para a ocorrência de quedas em agosto
      
**2011** - Nenhum motivo claro foi o gatilho para essa queda.

**2014** - Ano de inverno cripto com inicio em dezembro de 2013, teve como gatilho inicial notícias sobre ações de 
repressão contra o Bitcoin pela china e continuação da queda ocasionada pela falência da corretora 'MTGox'.
*fonte:* 
[cnbc](https://www.cnbc.com/2013/12/18/bitcoin-price-halves-as-china-clampdown-escalates.html)
[elpais](https://brasil.elpais.com/brasil/2014/02/25/tecnologia/1393319825_988195.html)

**2015** - Final do inverno em setembro, inicio da reversão do cíclo.

**2016** - Continuação do movimento de alta, leve baixa normal em março, preço despenca em mais de 20% em agosto 
por conta de ataque hacker na corretora 'bitfinex', muito popular na época. Em setembro o preço volta a subir.
*fonte:* 
[BBC](https://www.bbc.com/news/technology-37009319)

**2018** - Inverno cripto com inicio em janeiro

**2019** - Recuo leve do preço ocasionado pela queda das bolsas mundiais após boatos e noticias sobre uma possível recessão
norte-americana.
"...a taxa de juros dos títulos do Tesouro norte-americano de 10 anos caiu temporariamente abaixo da taxa de juros da
dívida de dois anos, pela primeira vez desde 2007, quando a economia norte-americana estava entrando na Grande Recessão.
Esse fenômeno conhecido como inversão da curva de rendimento, uma alteração da relação normal entre os títulos é geralmente 
considerado um sinal de recessão próxima."
*fonte:*
[elpais](https://brasil.elpais.com/brasil/2019/08/14/economia/1565817807_019685.html)

**2022** - Ano de bearmarket iniciado em novembro de 2021. Eventos macroeconômicos e ligados ao mercado de criptomoedas foram
o gatilho.

**2023** - Queda normal ocasionada pelo aumento da taxa de juros americana e noticias sobre a venda de bitcoins pela spaceX
*fonte:*
[livecoins](https://livecoins.com.br/bitcoin-cai-e-assusta-ate-os-investidores-mais-confiantes/)
[investnews](https://investnews.com.br/financas/reducao-de-bitcoin-da-spacex-elon-musk-parou-de-confiar-na-moeda/)

**2024** - O FED (Federal Reserve System) manteve a taxa de juros alta.


**Outras fontes:**
[InfoMoney](https://www.infomoney.com.br/mercados/sell-in-may-and-go-away-o-que-significa-esse-velho-ditado-do-mercado/)
```
